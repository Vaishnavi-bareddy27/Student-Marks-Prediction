```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
hours = [1, 2, 3, 4.5, 5, 6.1, 7, 8.5, 9, 10]
marks = [10, 20, 27, 45, 50, 62, 70, 85, 90, 95]

df = pd.DataFrame({"Hours": hours, "Marks": marks})
print(df)
```

```python
X = df[["Hours"]]   
y = df["Marks"]     
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Test Data:")
print(pd.DataFrame({"Hours": X_test["Hours"], "Actual Marks": y_test, "Predicted Marks": y_pred.round(2)}))
```

```python
plt.scatter(df["Hours"], df["Marks"], color="blue", label="Data points")
plt.plot(df["Hours"], model.predict(df[["Hours"]]), color="red", label="Regression line")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.title("Student Marks Prediction")
plt.legend()
plt.show()
```

```python
hours_studied =int(input("Enter hours studied: "))
predicted_marks = model.predict([[hours_studied]])
predicted_marks = min(predicted_marks, 100)
print(f"If a student studies {hours_studied} hours → Predicted Marks: {predicted_marks:.2f}")
```
